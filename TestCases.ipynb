```python
def Temp_Regime_1(x):
    val_T_Reg1 = (1.06e-9 * x**3) - (4.251e-5 * x**2) + (1.435 * x) - 85.42
    return val_T_Reg1

def Temp_Regime_2(x):
    val_T_Reg2 = (7.327e-9 * x**3) - (3.412e-4 * x**2) + (4.954 * x) - 943.9
    return val_T_Reg2

def Press_Regime_1(x):
    val_P_Reg1 = (2.879e-9 * x**3) - (1.847e-4 * x**2) + (7.401 * x) - 49.0
    return val_P_Reg1

def Press_Regime_2(x):
    val_P_Reg2 = (5.873e-14 * x**5) - (3.62e-9 * x**4) + (8.115e-5 * x**3) - (0.7849 * x**2) + (3105 * x) - 2.892e4
    return val_P_Reg2

def MomMag_Regime_1(x):
    val_M_Reg1 = (-0.206 * x**3) + (2.325 * x**2) - (0.9151 * x) + 6.964
    return val_M_Reg1

def MomMag_Regime_2(x):
    val_M_Reg2 = (-58.27 * x**3) + (368.8 * x**2) - (161.7 * x) + 318.9
    return val_M_Reg2

def VelMag_Regime_1(x):
    val_V_Reg1 = (6.521e-9 * x**3) - (4.757e-5 * x**2) + (0.9593 * x) + 294.9
    return val_V_Reg1

def VelMag_Regime_2(x):
    val_V_Reg2 = (9.836e-8 * x**3) - (1.04e-3 * x**2) + (3.331 * x) - 750.9
    return val_V_Reg2
```

```python
def mag_calc_vel(df):
    mag_list_vel = []
    for i in range(0, len(df)):
        v_x = df['Velocity Mag:0']
        v_y = df['Velocity Mag:1']
        v_z = df['Velocity Mag:2']
        magnitude = np.sqrt(v_x[i]**2 + v_y[i]**2 + v_z[i]**2)
        print(magnitude)
        mag_list_vel.append(magnitude)
        
    return mag_list_vel

def mag_calc_mom(df):
    mag_list_mom = []
    for i in range(0, len(df)):
        m_x = df['Momentum:0']
        m_y = df['Momentum:1']
        m_z = df['Momentum:2']
        magnitude = np.sqrt(m_x[i]**2 + m_y[i]**2 + m_z[i]**2)
        print(magnitude)
        mag_list_mom.append(magnitude)
        
    return mag_list_mom
```
