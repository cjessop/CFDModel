```python
def polyfit(x, y, degree):
    results = {}
    coeffs = np.polyfit(x, y, degree)
    p = np.poly1d(coeffs)
    #calculate r-squared
    yhat = p(x)
    ybar = np.sum(y)/len(y)
    ssreg = np.sum((yhat-ybar)**2)
    sstot = np.sum((y - ybar)**2)
    results['r_squared'] = ssreg / sstot

    return results

#find r-squared of polynomial model with degree = 3
print(polyfit(T1, T2, 3))
print(polyfit(T1, T3, 3))
print(polyfit(T1, T4, 3))
print(polyfit(T1, T5, 3))
```

```python
av_val_regime1 = []
for i in range(0, len(T1)):
    av_value_regime1 = (T2[i] + T3[i] + T4[i] + T5[i]) / 4
    av_val_regime1.append(av_value_regime1)
print(av_val_regime1)

av_val_regime2 = []
for i in range(0, len(T1)):
    av_value_regime2 = (T7[i] + T8[i]) / 2
    av_val_regime2.append(av_value_regime2)
print(av_val_regime2)
```

```python
model_av_regime1 = np.poly1d(np.polyfit(T1, av_val_regime1, 3))
print(model_av_regime1)
model_av_regime1_str = str(model_av_regime1)
print(model_av_regime1_str)
polyline_av_regime1 = np.linspace(1, 27000, 50)
plt.scatter(T1, av_val_regime1, alpha=0.8, s=9)
plt.plot(polyline_av_regime1, model_av_regime1(polyline_av_regime1), color='r')
plt.xlabel('Run 1 Temperature [K]')
plt.ylabel('Run Averaged Temperature [K]')
plt.legend(["Average Fitting Function"], loc='upper left', prop={'size': 16})
plt.annotate(model_av_regime1_str, xy =(3.3, 1),
                xytext =(6000, 3000), fontsize = 12  
                )
plt.title('Average Temperature Function Regime 1')
#plt.savefig('TvT_prime_Function_regime1.png', dpi=1200, format='png', bbox_inches='tight')
plt.show()
```

```python
model_av_regime2 = np.poly1d(np.polyfit(T1, av_val_regime2, 3))
print(model_av_regime2)
model_av_regime2_str = str(model_av_regime2)
print(model_av_regime2_str)
polyline_av_regime2 = np.linspace(1, 27000, 50)
plt.scatter(T1, av_val_regime2, alpha=0.8, s=9)
plt.plot(polyline_av_regime2, model_av_regime2(polyline_av_regime2), color='r')
plt.xlabel('Run 1 Temperature [K]')
plt.ylabel('Run Averaged Temperature [K]')
plt.legend(["Average Fitting Function"], loc='upper left', prop={'size': 16})
plt.annotate(model_av_regime1_str, xy =(3.3, 1),
                xytext =(6000, 3000), fontsize = 12  
                )
plt.title('Average Temperature Function Regime 2')
plt.savefig('TvT_prime_Function_regime2.png', dpi=1200, format='png', bbox_inches='tight')
plt.show()
```
