```python
def mag_calc(df):
    mag_list = []
    for i in range(0, len(df)):
        v_x = df['Velocity Mag:0']
        v_y = df['Velocity Mag:1']
        v_z = df['Velocity Mag:2']
        magnitude = np.sqrt(v_x[i]**2 + v_y[i]**2 + v_z[i]**2)
        print(magnitude)
        mag_list.append(magnitude)
        
    return mag_list
```

```python
av_val_regime1 = []
for i in range(0, len(v1)):
    av_value_regime1 = (v2[i] + v3[i] + v4[i] + v5[i] + v6[i]) / 5
    av_val_regime1.append(av_value_regime1)
print(av_val_regime1)

av_val_regime2 = []
for i in range(0, len(v1)):
    av_value_regime2 = (v7[i] + v8[i] + v9[i] + v10[i] + v11[i]) / 5
    av_val_regime2.append(av_value_regime2)
print(av_val_regime2)
```

```python
model_av_regime1 = np.poly1d(np.polyfit(v1, av_val_regime1, 3))
print(model_av_regime1)
model_av_regime1_str = str(model_av_regime1)
print(model_av_regime1_str)
polyline_av_regime1 = np.linspace(1, 8000, 50)
plt.scatter(v1, av_val_regime1, alpha=0.8, s=9)
plt.plot(polyline_av_regime1, model_av_regime1(polyline_av_regime1), color='r')
plt.xlabel('Run 1 Velocity Magnitude [m/s]')
plt.ylabel('Run Averaged Velocity Magnitude [m/s]')
plt.legend(["Average Fitting Function"], loc='upper left', prop={'size': 16})
plt.annotate(model_av_regime1_str, xy =(3.3, 1),
                xytext =(1000, 500), fontsize = 12  
                )
plt.title('Average Velocity Mag Function Regime 1')
#plt.savefig('VvV_prime_Function_regime1.png', dpi=1200, format='png', bbox_inches='tight')
plt.show()
```
